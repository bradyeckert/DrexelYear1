```python
import matplotlib.pyplot as plt
import numpy as np

xgrid = np.arange(0.0,1.0, 0.0005)
plt.figure(figsize=(10,10))

for N in [1,5,10,20,50,100,500]:
    S=xgrid*0
    for i in range (1,N+1):
        bi=((np.pi*i-np.sin(np.pi*i))/((np.pi*i)**2))
        f = 2*bi*(np.sin(i*np.pi*xgrid))
        S=S+f

    plt.plot(xgrid,S)
plt.show
result=np.where(S==np.amax(S))
#print(max(S),result[0])
print('The maximum value of S when N=500 is ',max(S),' so the overshoot is ',max(S)-1.,'\n')
print('The location of the maximum value of S when N=500 is at x=',xgrid[4], ' or 1/500 or 1/N')

```
